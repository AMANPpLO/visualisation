```python
### 3D Surface Plots
```

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
```

```python
xx, yy = np.meshgrid(x,y)
```

```python
z = xx**2 + yy**2
z.shape
```

```python
fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)#fig ko variable me dal ka usko fig me dalna hoga tavi kam ka
```

```python
z = np.sin(xx) + np.cos(yy)

fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
z = np.sin(xx) + np.log(xx)

fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)#sir na bas asa he dikhaya hai bro ok na 
```

```python
z = np.cos(xx) + np.cos(xx)
l = np.sin(xx) + np.sin(xx)
fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx,yy,z,cmap='viridis')
q = ax.plot_surface(xx,yy,l,cmap='winter')
fig.colorbar(p)#sir na bas asa he dikhaya hai bro ok na 
fig.colorbar(q)
```

### Heatmap

```python
delivery =pd.read_csv('D:\download\python\IPL_Ball_by_Ball_2008_2022.csv')
delivery.head()
```

```python
temp_df = delivery[(delivery['ballnumber'].isin([1,2,3,4,5,6])) & (delivery['batsman_run']==6)]
```
